Randomly selecting 10K rows from the english and portuguese songs, in order to create smaller datasets. 



```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pickle
import pandas as pd
```

English

---



```python
with open ('/content/drive/MyDrive/Progetto_TA_2022/pickled_df/df_en.pkl', mode='br') as inputfile:
  df_en = pickle.load(inputfile)
```

```python
df_en_10K = df_en.sample(n=10000)
```

```python
df_en_10K.shape
```

```python
df_en_10K = df_en_10K.reset_index()
df_en_10K = df_en_10K.drop(['index'], axis = 1)
```

```python
df_en_10K.head()
```

```python
with open('/content/drive/MyDrive/Progetto_TA_2022/pickled_df/df_en_10K.pkl','wb') as outfile:
    pickle.dump(df_en_10K,outfile)
```
